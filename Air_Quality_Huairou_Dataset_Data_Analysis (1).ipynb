## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
data = pd.read_csv("PRSA_Data_Huairou_20130301-20170228.csv")
```

### Assessing Data

```python
# Check the first few rows of the dataframe
print(data.head())

# Get a summary of the dataframe
print(data.info())

# Get descriptive statistics
print(data.describe())
```

### Cleaning Data

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)

# Drop rows with missing values
data_cleaned = data.dropna()

# Verify that there are no more missing values
print(data_cleaned.isnull().sum())
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Convert date columns to datetime
data_cleaned['year'] = data_cleaned['year'].astype(str)
data_cleaned['month'] = data_cleaned['month'].astype(str)
data_cleaned['day'] = data_cleaned['day'].astype(str)
data_cleaned['hour'] = data_cleaned['hour'].astype(str)

data_cleaned['datetime'] = pd.to_datetime(data_cleaned['year'] + '-' + data_cleaned['month'] + '-' + data_cleaned['day'] + ' ' + data_cleaned['hour'] + ':00:00')

# Set datetime as index
data_cleaned.set_index('datetime', inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Plot PM2.5 over time
plt.figure(figsize=(12, 6))
data_cleaned['PM2.5'].resample('M').mean().plot()
plt.title('Monthly Average PM2.5 in Huairou (2013-2017)')
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.show()
```

### Pertanyaan 2:

```python
# Plot PM2.5 by month
plt.figure(figsize=(12, 6))
sns.boxplot(x=data_cleaned.index.month, y=data_cleaned['PM2.5'])
plt.title('Monthly Distribution of PM2.5 in Huairou')
plt.xlabel('Month')
plt.ylabel('PM2.5')
plt.show()
```

# Pertanyaan 3

```python
# Plot PM2.5 against temperature
plt.figure(figsize=(12, 6))
sns.scatterplot(x=data_cleaned['TEMP'], y=data_cleaned['PM2.5'])
plt.title('PM2.5 vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('PM2.5')
plt.show()

# Plot PM2.5 against pressure
plt.figure(figsize=(12, 6))
sns.scatterplot(x=data_cleaned['PRES'], y=data_cleaned['PM2.5'])
plt.title('PM2.5 vs Pressure')
plt.xlabel('Pressure')
plt.ylabel('PM2.5')
plt.show()

# Plot PM2.5 against wind speed
plt.figure(figsize=(12, 6))
sns.scatterplot(x=data_cleaned['WSPM'], y=data_cleaned['PM2.5'])
plt.title('PM2.5 vs Wind Speed')
plt.xlabel('Wind Speed')
plt.ylabel('PM2.5')
plt.show()
```

## Conclusion

**Pertanyaan 1**
Tren Kualitas Udara (PM2.5) dari Tahun 2013 hingga 2017:
*   Tingkat PM2.5 mengalami fluktuasi yang signifikan selama periode analisis. Terdapat beberapa puncak konsentrasi PM2.5 yang tinggi pada waktu-waktu tertentu dalam setahun, serta penurunan yang signifikan pada waktu lainnya.
*   Ada variasi tahunan dalam kualitas udara, yang menunjukkan bahwa ada periode-periode tertentu dengan kualitas udara yang lebih buruk. Upaya pengendalian polusi udara perlu ditingkatkan selama periode ini untuk mengurangi dampak negatif pada kesehatan masyarakat.

**Pertanyaan 2**
Pola Musiman dalam Tingkat PM2.5:
*   Pola musiman terlihat jelas dalam data, dengan tingkat PM2.5 cenderung lebih tinggi selama musim dingin dan lebih rendah selama musim panas.
*   Faktor musiman berpengaruh signifikan terhadap tingkat polusi udara. Selama musim dingin, kondisi atmosfer mungkin menyebabkan akumulasi polutan, sementara kondisi selama musim panas mungkin lebih kondusif untuk dispersi polutan.

**Pertanyaan 3**
Pengaruh Faktor Meteorologi terhadap Tingkat PM2.5:
> Suhu (TEMP):
*   Ada kecenderungan bahwa PM2.5 lebih tinggi pada suhu yang lebih rendah.
*   Suhu yang lebih rendah mungkin berkontribusi pada peningkatan konsentrasi PM2.5, kemungkinan karena pembakaran bahan bakar untuk pemanasan selama musim dingin.

> Tekanan (PRES):
*   Hubungan antara tekanan dan PM2.5 tidak menunjukkan pola yang sangat jelas, tetapi ada beberapa variasi yang bisa diamati.
*   Tekanan atmosfer mungkin berinteraksi dengan faktor lain untuk mempengaruhi konsentrasi PM2.5, tetapi tidak ada pola tunggal yang jelas dari data ini.

> Kecepatan Angin (WSPM):
*   Tingkat PM2.5 cenderung lebih rendah pada kecepatan angin yang lebih tinggi.
*   Angin yang lebih kencang membantu mendispersikan polutan di udara, mengurangi konsentrasi PM2.5. Ini menunjukkan pentingnya sirkulasi udara yang baik untuk menjaga kualitas udara.
